Firstly, let us calculate the wigner distribution for different values of $\theta$ and $\phi$, where 
$\theta$ and $\phi$ are the coordinates on the Bloch sphere. An alternative representation for a two-level atom can be given by the Wigner function. For a two-level system, we need to introduce a generalized parity operator and a rotation operator which are paramaterized by SU(2). The rotation operator we use is standard Euler angles, such that

$U(\theta,\phi, \Phi) = \exp\left(\mathrm{i}\sigma_z\phi\right) \exp\left(\mathrm{i}\sigma_y\theta\right) \exp\left(\mathrm{i}\sigma_z\Phi\right)$

and the parity is

$\Pi = \frac{1}{2}\left(\mathbb{1}+\sqrt{3}\sigma_z\right)$

where $\sigma_y$ and $\sigma_z$ are the standard Pauli matrices. The kernel for the Wigner function is then defined as

$\Pi(\theta,\phi) = U(\theta,\phi, \Phi) \Pi U^\dagger(\theta,\phi, \Phi)$

Where the $\Phi$ degrees of freedom cancel out. The Wigner function is now generated by

$W(\theta,\phi) = \mathrm{Tr}\left[\rho \, \Pi(\theta,\phi) \right],$

where $\rho$ is the density matrix for the state. The kernel can be extended to be used for any number of two-level atoms as

$\Pi(\boldsymbol{\theta},\boldsymbol{\phi}) = \bigotimes_{i=1}^N \Pi(\theta_i,\phi_i),$

for $N$ atoms, where $\boldsymbol{\theta} = [\theta_1, ..., \theta_2]$ and likewise for $\boldsymbol{\phi}$.

As an example, we choose the two-atom, maximally entangled Bell state $|\rm \uparrow \rm \uparrow\rangle + |\rm \downarrow \rm \downarrow \rangle $. We set the total angular momentum number ($j$) to 0.5, as there are just two states: $\left|\rm up \right>$ and $\left|\rm down \right>$. Since our state is for a two particle system, we have the choice of using a tensor product of two SU(2) parities or the full SU(4) parity, which we can set using the fullparity Boolean. The number of theta and phi points at which the wigner transformation is to be calculated can be set using the steps parameter. Finally, slicearray is an array of strings, each one either being 'l', 'x', 'y' or 'z' to select the equal, X, Y or Z angle slice respectively.

```python
% matplotlib inline
from qutip.states import bell_state
from qutip.wigner import wigner_transform
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from qutip.visualization import plot_wigner_sphere

psi = bell_state('00')
j = 0.5
sun = 2
fullparity = False
steps = 100
slicearray = ['l', 'l']
wigner = wigner_transform(psi, j, fullparity, steps, slicearray)
```

The ComplexWarning can be ignored; it is caused by rounding errors. Now that we have the wigner distribution, we can plot it. We create a pyplot figure and a new axes instance, and set the reflections Boolean to false to see just the Bloch sphere:

```python
fig = plt.figure()
ax = Axes3D(fig)
reflections = False
plot_wigner_sphere(fig, ax, wigner, reflections)
```
