Basic 2D Plotting
-----------------
Before generating plots in Python we must first load the main Matplotlib module.

````python
import matplotlib.pyplot as plt
````

Let us also import numpy so we can use arrays and mathematical functions

````python
import numpy as np
````

### Drawing Lines
Plotting a simple function, such as a sine function, is easy to do.  All we need are two arrays, one for the x values, and one for the f(x) values

````python
x = np.linspace(-np.pi, np.pi)
y = np.sin(x)
plt.plot(x, y)
plt.grid()      # Adds grids to plot
plt.show()
````

Here, the plot command generates the figure, but it is not displayed until you run ```plt.show()```.  If we want, we can also also add some labels to the axes and a title to the plot. While we are at it, lets change the color of the line to red, and make it a dashed line.

````python
x = np.linspace(-np.pi, np.pi)
y = np.sin(x)
plt.plot(x, y, 'r--') # Make line red 'r' and dashed '--'
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin(x)')
plt.grid()
plt.show()
````

Here the 'r' stands for red, but we could have used any of the built in colors:

- 'b': blue
- 'g': green
- 'r': red
- 'c': cyan
- 'm': magenta
- 'y': yellow
- 'k': black
- 'w': white

We can also specify the color of a line using the color keyword argument

````python
x = np.linspace(-np.pi, np.pi)
y = np.sin(x)
plt.plot(x, y, '--', color='0.75') # Here a string from 0->1 specifies a gray value.
plt.grid()
plt.show()
````

````python
x = np.linspace(-np.pi, np.pi)
y = np.sin(x)
plt.plot(x, y, '-', color='#FD8808') # We can also use hex colors if we want.
plt.grid()
plt.show()
````

The style of the line can be changed from solid: ```''``` or ```'-'```, to dashed: ```'--'```, dotted ```'.'```, dash-dotted: ```'-.'```, dots+solid: ```'.-'```, or little dots: ```':'```.  One can also use the 'linstyle' or 'ls' keyword argument to change this style.  We can disply all of these variations using the **subplot** function that displays multiple plots in a grid specified by the number of rows, columns, and the number of the current plot.  We only need one ```plt.show()``` command for viewing all of the plots.  To make the plot look good, we can also control the width and height of the figure by calling the ```plt.figure``` function using the keyword argument 'figsize' that specifies the width and height of the figure in inches.

````python
x = np.linspace(-np.pi, np.pi)
y = np.sin(x)

plt.figure(figsize=(12, 6)) # This controls the size of the figure

plt.subplot(2, 3, 1) # This is the first plot in a 2x3 grid of plots
plt.plot(x, y)
plt.grid()

plt.subplot(2, 3, 2)  # This is the second plot
plt.plot(x, y, linestyle='--') # Demo using 'linestyle' keyword arguement
plt.grid()

plt.subplot(2, 3, 3)
plt.plot(x, y, '.')
plt.grid()

plt.subplot(2, 3, 4)
plt.plot(x, y, '-.')
plt.grid()

plt.subplot(2, 3, 5)
plt.plot(x, y, '.-')
plt.grid()

plt.subplot(2, 3, 6)
plt.plot(x, y, ls=':') # Demo using 'ls' keyword arguement.
plt.grid()

plt.show()
````

If we want to change the width of a line then we can use the 'linewidth' or 'lw' keyword arguements with a float number specifying the linewidth

````python
x = np.linspace(0, 10)
y = np.sqrt(x)

plt.figure(figsize=(12, 3))

plt.subplot(1, 3, 1)
plt.plot(x, y)
plt.grid()

plt.subplot(1, 3, 2)
plt.plot(x, y, linewidth=2)
plt.grid()

plt.subplot(1, 3, 3)
plt.plot(x, y, lw=7.75)
plt.grid()

plt.show()
````

If we want to plot multiple lines on a single plot, we can call the plot command several times, or we can use just a single plot command by entering the data for multiple lines simultaneously.

````python
x = np.linspace(0, 10)
s = np.sin(x)
c = np.cos(x)
sx = x * np.sin(x)

plt.figure(figsize=(12,3))

plt.subplot(1, 2, 1)
plt.plot(x, s, 'b')    # Call 3 different plot functions
plt.plot(x, c, 'r')
plt.plot(x, sx, 'g')
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(x, s, 'b', x, c, 'r', x, sx, 'g') # Combine multiple lines in one call to plot
plt.grid()

plt.show()
````

### Setting Axes Limits
Suppose we want to limit the range of either the x-axis or y-axis, or both.  In Matplotlib we can use the ```plt.xlim``` and ```plt.ylim``` functions to accomplish this.

````python
x = np.linspace(-np.pi, np.pi,100)
plt.figure(figsize=(12,3))

plt.subplot(1, 3, 1)
plt.plot(x, np.sin(x), lw=2)
plt.grid()

plt.subplot(1, 3, 2)
plt.plot(x, np.sin(x), lw=2, color='#740007')
plt.xlim([-np.pi, np.pi])                       # Change bounds on x-axis to [-pi,pi]
plt.grid()

plt.subplot(1, 3, 3)
plt.plot(x, np.sin(x), '^', ms=8, color='0.8')
plt.xlim([-1, 1])                         # Change bounds on x-axis to [-1,1]
plt.ylim([-0.75, 0.75])                   # Change bounds on y-axis to [-0.75,0.75]
plt.grid()

plt.show()
````

### Saving Figures
Now that we know how to make good looking plots, it would certainly be nice if we knew how to save these figures for use in a paper/report, or perhaps for posting on a webpage.  Fortunately, this is very easy to do.  If we want to save our previous figure then we need call the ```plt.savefig``` function.

````python
x = np.linspace(-np.pi, np.pi, 100)

plt.figure(figsize=(12, 3))
plt.subplot(1, 3, 1)
plt.plot(x, np.sin(x), lw=2)
plt.grid()

plt.subplot(1, 3, 2)
plt.plot(x, np.sin(x), lw=2, color='#740007')
plt.xlim([-np.pi, np.pi])                       
plt.grid()

plt.subplot(1,3,3)
plt.plot(x, np.sin(x), '^', ms=8, color='0.8')
plt.xlim([-1, 1])                         
plt.ylim([-0.75, 0.75])                   
plt.grid()

plt.savefig('axes_example.png') # Save the figure in PNG format in same directory as script
````

```plt.savefig``` saves the figure with the name and extension that is given in the string.  The name can be whatever you like, but the extension. ```.png``` in this case, must be a format that Matplotlib recognizes.  In this class we will only use the Portable Network Graphics ```.png``` and PDF ```.pdf``` formats.

### Using Shapes for Data Points
In addition to lines and dots, Matplotlib allows you to use many different shapes to represent the points on a graph.  In Matplotlib, these shapes are called **markers** and just like lines, their color and size can be controlled.  There are many [basic types of markers](http://matplotlib.org/api/markers_api.html), so here we will demonstrate just a few important ones: '*': star, 'o': circle, 's': square, and '+': plus by evaluating 
the Airy functions.

````python
from scipy.special import airy # Airy functions are in the SciPy special module
x = np.linspace(-1, 1)
Ai, Aip, Bi, Bip = airy(x)
plt.plot(x, Ai, 'b*', x, Aip, 'ro', x, Bi, 'gs', x, Bip, 'k+')
plt.grid()
plt.show()
````

We can also chage the size of the shapes using the 'markersize' or 'ms' keyword arguments, and the color using 'markercolor' or 'mc'.

Other Kinds of 2D Plots
-----------------------

So far we have made use of only the ```plot``` function for generating 2D figures.  However, there are several other functions for generating different kinds of 2D plots. A collection of the many different types can be found at the [matplotlib gallery](http://matplotlib.org/gallery.html), and we will highlight only the more useful functions.

````python
x = np.linspace(-1, 1., 100)

plt.figure(figsize=(6, 6))
plt.subplot(2, 2, 1)
y = x + 0.25*np.random.randn(len(x))
plt.scatter(x, y, color='r') # Plot collection of (x,y) points
plt.title('A scatter plot')
plt.grid()

plt.subplot(2, 2, 2)
n = np.array([0, 1, 2, 3, 4, 5])
plt.bar(n, n**2, align="center", width=1) # Aligns the bars over the x-numbers, and width=dx
plt.title('A bar plot')
plt.grid()

plt.subplot(2, 2, 3)
plt.fill_between(x, x**2, x**3, color="green") # Fill between x**2 & x**3 with green
plt.title('A fill-between plot')
plt.grid()

plt.subplot(2, 2, 4)
plt.title('A hist plot')

r = np.random.randn(50)    # Generating some random numbers
plt.hist(r, color='y')      # Create a histogram of the random number values
plt.grid()

plt.show()
````

The color and size of the elements can all be controlled in the same way as the usual ```plot``` function.
